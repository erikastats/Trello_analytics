```python
import os
import requests
import pandas as pd 
import numpy as np
from dotenv import load_dotenv
import json
import re
```

```python
load_dotenv()
```

```python
# Getting the variables 
API_KEY = os.getenv('TRELLO_API_KEY')
TOKEN = os.getenv('TRELLO_TOKEN')
BOARD_ID = os.getenv('TRELLO_BOARD_ID')
```

Board cards df

```python
# Getting board cards
url_cards = f"https://api.trello.com/1/boards/{BOARD_ID}/cards?key={API_KEY}&token={TOKEN}"
cards = requests.get(url_cards).json()
```

```python
cards[0].keys()
```

```python
'id', 'badges', 'closed','due', 'dueComplete', 'dateLastActivity', 'labels'
```

```python
df = pd.json_normalize(cards)
print(df.columns.tolist())
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head(2)
```

```python
columns_selected = ['id', 'dateLastActivity', 'desc', 'due', 'name']
cards_selected = df[columns_selected]
cards_selected.head(2)
```

Action history

```python
card_id = cards_selected['id']
card_id
```

```python
actions_df = pd.DataFrame()
for id in card_id:
    url_actions = f"https://api.trello.com/1/cards/{id}/actions?filter=updateCard&key={API_KEY}&token={TOKEN}"
    actions = requests.get(url_actions).json()
    df_a = pd.json_normalize(actions)
    # df_a = df_a[['id','date', 'data.card.name', 'data.card.id', 'data.list.name']]
    actions_df = pd.concat([actions_df, df_a], ignore_index= True)
```

```python
actions_df = actions_df[['id','date', 'data.card.name', 'data.card.id', 'data.list.name']]
```

```python
actions_df.head(2)
```

Get comments

```python
comments_df = pd.DataFrame()
for id in card_id:
    url_actions =  f"https://api.trello.com/1/cards/{id}/actions?filter=commentCard&key={API_KEY}&token={TOKEN}"
    actions = requests.get(url_actions).json()
    df_a = pd.json_normalize(actions)
    # df_a = df_a[['id','date', 'data.card.name', 'data.card.id', 'data.list.name']]
    comments_df = pd.concat([comments_df, df_a], ignore_index= True)
```

```python
comments_columns = ['id', 'date', 'data.card.id', 'data.text']
```

```python
comments_df = comments_df[comments_columns]
```

Cleaning the data


```python
def cleaning_text(data, column_name):
    data[column_name] = data[column_name].apply(lambda x: re.sub(r'\s+', ' ', x.strip()).lower())
    data[column_name] = data[column_name].apply(lambda x : re.sub(r"[#'*]", '', x))
    return data
```

```python
comments_df_cleaned = cleaning_text(comments_df, 'data.text')
cards_selected_cleaned = cleaning_text(cards_selected, 'desc')
actions_df_cleaned = cleaning_text(actions_df, 'data.card.name')
```

```python
cards_selected_cleaned.head(2)
```

```python
comments_df_cleaned.to_csv("comments_trello.csv", index = False)
cards_selected_cleaned.to_csv("cards_trello.csv", index = False)
actions_df_cleaned.to_csv("actions_trello.csv", index = False)
```
